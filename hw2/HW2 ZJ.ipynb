```python
def add_features_to_df(df, key):
    # Define column names based on the key
    open_col = f'{key}_Open'
    high_col = f'{key}_High'
    low_col = f'{key}_Low'
    close_col = f'{key}_Close'
    volume_col = f'{key}_Volume'
    
    # Ensure data is sorted chronologically
    df = df.sort_index()
    
    # Calculate percentage returns
    df[f'{key}_Return'] = df[close_col].pct_change()
    
    # Calculate cumulative returns
    df[f'{key}_Cumulative_Return'] = (1 + df[f'{key}_Return']).cumprod() - 1
    
    # Calculate rolling statistics on returns
    for window in [3, 5, 10, 20]:
        df[f'{key}_Return_SMA_{window}'] = df[f'{key}_Return'].rolling(window=window, min_periods=1).mean()
        df[f'{key}_Return_STD_{window}'] = df[f'{key}_Return'].rolling(window=window, min_periods=1).std()
        df[f'{key}_Cumulative_Return_{window}'] = (1 + df[f'{key}_Return']).rolling(window=window, min_periods=1).apply(lambda x: np.prod(x) - 1, raw=False)
    
    # Calculate price differences
    df[f'{key}_Open_Close_Diff'] = df[open_col] - df[close_col]
    df[f'{key}_High_Low_Diff'] = df[high_col] - df[low_col]
    
    # Simple Moving Averages and Standard Deviations of price
    for window in [3, 5, 10, 20]:
        df[f'{key}_SMA_{window}'] = df[close_col].rolling(window=window, min_periods=1).mean()
        df[f'{key}_STD_{window}'] = df[close_col].rolling(window=window, min_periods=1).std()
        df[f'{key}_Volume_SMA_{window}'] = df[volume_col].rolling(window=window, min_periods=1).mean()
    
    # Relative Strength Index (RSI)
    delta = df[close_col].diff()
    up = delta.clip(lower=0)
    down = -delta.clip(upper=0)
    window_length = 14
    roll_up = up.rolling(window_length).mean()
    roll_down = down.rolling(window_length).mean()
    rs = roll_up / roll_down
    df[f'{key}_RSI_{window_length}'] = 100.0 - (100.0 / (1.0 + rs))
    
    # Moving Average Convergence Divergence (MACD)
    ema_12 = df[close_col].ewm(span=12, adjust=False).mean()
    ema_26 = df[close_col].ewm(span=26, adjust=False).mean()
    df[f'{key}_MACD'] = ema_12 - ema_26
    df[f'{key}_Signal_Line'] = df[f'{key}_MACD'].ewm(span=9, adjust=False).mean()
    
    # Rate of Change (ROC)
    df[f'{key}_ROC'] = df[close_col].pct_change(periods=window_length) * 100
    
    # Lagged Returns (for capturing momentum)
    for lag in [1, 2, 3, 5, 10]:
        df[f'{key}_Return_Lag_{lag}'] = df[f'{key}_Return'].shift(lag)
    
    # Ensure no lookahead bias by not using future data in features
    # All features are calculated using past and current data only
    
    return df

# Assuming 'indicator_dfs' is your dictionary of dataframes and 'indicators' is the list of keys
for key in indicators:
    df = indicator_dfs[key]
    df = add_features_to_df(df, key)
    indicator_dfs[key] = df
```

```python
def add_features_to_crypto_df(df, key):
    # Define column names
    open_col = f'{key}_Open'
    high_col = f'{key}_High'
    low_col = f'{key}_Low'
    close_col = f'{key}_Close'
    volume_col = f'{key}_Volume'
    trades_col = f'{key}_Trades'
    
    # Ensure data is sorted chronologically
    df = df.sort_index()
    
    # Calculate percentage returns
    df[f'{key}_Return'] = df[close_col].pct_change()
    
    # Calculate cumulative returns
    df[f'{key}_Cumulative_Return'] = (1 + df[f'{key}_Return']).cumprod() - 1
    
    # Calculate rolling statistics on returns over past 60 and 90 minutes
    for window in [30, 60, 90]:
        df[f'{key}_Return_SMA_{window}'] = df[f'{key}_Return'].rolling(window=window, min_periods=1).mean()
        df[f'{key}_Return_STD_{window}'] = df[f'{key}_Return'].rolling(window=window, min_periods=1).std()
        df[f'{key}_Cumulative_Return_{window}'] = (1 + df[f'{key}_Return']).rolling(window=window, min_periods=1).apply(lambda x: np.prod(x) - 1, raw=False)
    
    # Calculate price differences
    df[f'{key}_Open_Close_Diff'] = df[open_col] - df[close_col]
    df[f'{key}_High_Low_Diff'] = df[high_col] - df[low_col]
    
    # Moving Averages and Standard Deviations of price over past 60 and 90 minutes
    for window in [30, 60, 90]:
        df[f'{key}_SMA_{window}'] = df[close_col].rolling(window=window, min_periods=1).mean()
        df[f'{key}_STD_{window}'] = df[close_col].rolling(window=window, min_periods=1).std()
        df[f'{key}_Volume_SMA_{window}'] = df[volume_col].rolling(window=window, min_periods=1).mean()
        df[f'{key}_Trades_SMA_{window}'] = df[trades_col].rolling(window=window, min_periods=1).mean()
    
    # Relative Strength Index (RSI) over past 60 and 90 minutes
    for window_length in [30, 60, 90]:
        delta = df[close_col].diff()
        up = delta.clip(lower=0)
        down = -delta.clip(upper=0)
        roll_up = up.rolling(window_length).mean()
        roll_down = down.rolling(window_length).mean()
        rs = roll_up / roll_down
        df[f'{key}_RSI_{window_length}'] = 100.0 - (100.0 / (1.0 + rs))
    
    # Moving Average Convergence Divergence (MACD) adjusted for minute data
    # Use spans that reflect the objective time horizon
    ema_short_span = 12  # You might adjust these spans based on your data frequency
    ema_long_span = 26
    df[f'{key}_EMA_{ema_short_span}'] = df[close_col].ewm(span=ema_short_span, adjust=False).mean()
    df[f'{key}_EMA_{ema_long_span}'] = df[close_col].ewm(span=ema_long_span, adjust=False).mean()
    df[f'{key}_MACD'] = df[f'{key}_EMA_{ema_short_span}'] - df[f'{key}_EMA_{ema_long_span}']
    df[f'{key}_Signal_Line'] = df[f'{key}_MACD'].ewm(span=9, adjust=False).mean()
    
    # Volume per Trade
    df[f'{key}_Volume_per_Trade'] = df[volume_col] / df[trades_col]
    
    # Lagged Returns (for capturing momentum)
    for lag in [1, 5, 10, 15, 30, 60, 90]:
        df[f'{key}_Return_Lag_{lag}'] = df[f'{key}_Return'].shift(lag)
    
    # Time of Day Features (if datetime index is present)
    if isinstance(df.index, pd.DatetimeIndex):
        df[f'{key}_Hour'] = df.index.hour
        df[f'{key}_Minute'] = df.index.minute
    
    # Ensure no lookahead bias by not using future data in features
    
    return df

BTC = add_features_to_crypto_df(BTC, 'BTC')
ETH = add_features_to_crypto_df(ETH, 'ETH')
```

```python
def add_target_variable_new(df, key, future_window, threshold=0.001):
    # Define column names
    close_col = f'{key}_Close'
    
    # Ensure data is sorted chronologically
    df = df.sort_index()
    
    # Compute the average future close price over the next 'future_window'
    df['Avg_Future_Close'] = df[close_col].shift(-(future_window-1)).rolling(window=future_window).mean()
    
    # Create the target variable based on the threshold
    # 1 if Avg_Future_Close < (1 - threshold) * current close
    # 0 otherwise
    df['Target'] = np.where(
        df['Avg_Future_Close'] < (1 - threshold) * df[close_col], 
        1, 
        0
    )
    
    # Drop rows with NaN in the target variable (due to insufficient future data)
    df = df.dropna(subset=['Target'])
    
    # Drop intermediate columns
    df = df.drop(columns=['Avg_Future_Close'])
    
    return df

crypto_df = add_target_variable_new(crypto_df, 'BTC', 30, threshold=0.002)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, roc_auc_score
from xgboost import DMatrix, train, plot_importance
from tqdm import tqdm
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Define evaluation function
def evaluate_model(y_true, y_pred):
    print("\nConfusion Matrix:")
    conf_matrix = confusion_matrix(y_true, y_pred)
    print(conf_matrix)
    
    print("\nClassification Report:")
    print(classification_report(y_true, y_pred, target_names=['Class 0', 'Class 1']))
    
    accuracy = accuracy_score(y_true, y_pred)
    print(f"Accuracy Score: {accuracy:.4f}")
    
    # ROC-AUC (Optional)
    auc = roc_auc_score(y_true, y_pred)
    print(f"ROC-AUC Score: {auc:.4f}")

# Update X and y (binary classification)
X = crypto_df[feature_cols]
y = crypto_df['Target']

# Split the data into training and testing sets based on time
split_index = int(0.7 * len(X))

X_train = X.iloc[:split_index]
X_test = X.iloc[split_index:]
y_train = y.iloc[:split_index]
y_test = y.iloc[split_index:]

# Prepare DMatrix for XGBoost
dtrain = DMatrix(X_train, label=y_train)
dtest = DMatrix(X_test, label=y_test)

# XGBoost parameters for binary classification
params = {
    'objective': 'binary:logistic',
    'max_depth': 6,
    'learning_rate': 0.1,
    'subsample': 0.7,
    'colsample_bytree': 0.8,
    'reg_alpha': 0.1,
    'reg_lambda': 1.0,
    'random_state': 42,
    'eval_metric': 'logloss'
}

# Define a custom progress bar
epochs = 100
progress_bar = tqdm(range(epochs), desc="Training Progress")

# Train the model with custom loop for progress tracking
evals_result = {}
booster = train(
    params,
    dtrain,
    num_boost_round=epochs,
    evals=[(dtest, 'test')],
    evals_result=evals_result,
    verbose_eval=False
)

# Update the progress bar
for epoch in range(epochs):
    progress_bar.update(1)
progress_bar.close()

# Predictions
y_pred_prob = booster.predict(dtest)  # Predicted probabilities
y_pred = (y_pred_prob > 0.5).astype(int)  # Convert probabilities to binary predictions

# Evaluate model performance
evaluate_model(y_test, y_pred)

# Feature importance (Optional)
plt.figure(figsize=(10, 8))
plot_importance(booster, max_num_features=20)
plt.title('Feature Importance')
plt.show()

# Optional: Plot learning curves
plt.figure(figsize=(10, 6))
plt.plot(evals_result['test']['logloss'], label='Test Log Loss')
plt.xlabel('Epochs')
plt.ylabel('Log Loss')
plt.title('Learning Curve')
plt.legend()
plt.show()
```

```python
def add_position_and_trade_columns(df):
    # Initialize variables
    initial_investment = 10_000_000 * 0.999  # Initial USD capital after transaction cost
    btc_balance = initial_investment / df.iloc[0]['BTC_Close']  # Initial BTC holdings
    usd_balance = 0.0  # Start with no USD cash
    current_state = 'long'  # Start in a long position (holding BTC)
    position_list = []
    trade_list = []
    time_since_sell = 0
    sell_price = None  # Price at which BTC was sold

    # Iterate over each row in the DataFrame
    for idx, row in df.iterrows():
        decision = row['Decision']
        btc_close = row['BTC_Close']
        trade = 0  # Default trade value (0 means no trade)

        if current_state == 'long':
            usd_balance = 0.0  # Ensure USD balance is zero when long BTC
            if decision == 1:
                # Sell BTC position
                usd_received = btc_balance * btc_close * 0.999  # Apply transaction cost
                usd_balance = usd_received
                btc_balance = 0.0  # No BTC holdings after selling
                current_state = 'short'
                sell_price = btc_close
                time_since_sell = 0
                trade = 1
            else:
                # Remain in long position
                pass
        elif current_state == 'short':
            btc_balance = 0.0  # Ensure BTC balance is zero when short BTC
            time_since_sell += 1
            if btc_close < 0.998 * sell_price:
                # Buy back BTC position early
                amount_to_invest = usd_balance  # Apply transaction cost
                btc_balance = amount_to_invest / btc_close
                usd_balance = 0.0  # All USD converted back to BTC
                current_state = 'long'
                trade = 1
            elif time_since_sell >= 90:
                # Forced to buy back after 60 minutes
                amount_to_invest = usd_balance * 0.999  # Apply transaction cost
                btc_balance = amount_to_invest / btc_close
                usd_balance = 0.0
                current_state = 'long'
                trade = 1
            else:
                # Remain in short position
                pass

        # Calculate current position value in USD
        position_value = btc_balance * btc_close + usd_balance
        position_list.append(position_value)
        trade_list.append(trade)

    # Add the new columns to the DataFrame
    df['Position'] = position_list
    df['Trade'] = trade_list

    return df
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import timedelta
import statsmodels.api as sm

def calculate_metrics(df, returns_column, benchmark_returns_column=None):
    """
    Calculate performance metrics for a given returns series.
    
    Parameters:
    - df: DataFrame containing the returns data.
    - returns_column: Column name for the strategy returns.
    - benchmark_returns_column: Column name for the benchmark returns (if any).
    
    Returns:
    - metrics: Dictionary of calculated metrics.
    """
    # Calculate cumulative returns
    df['Cumulative_Returns'] = (1 + df[returns_column]).cumprod()
    
    # Calculate total time period in years
    total_days = (df.index[-1] - df.index[0]).total_seconds() / (3600 * 24)
    total_years = total_days / 365.25  # Accounting for leap years
    
    ## Metrics Calculation ##
    
    # Total Return
    total_return = df['Cumulative_Returns'].iloc[-1] - 1
    annualized_return = (1 + total_return) ** (1 / total_years) - 1
    
    # Number of Transactions (if 'Trade' column exists)
    if 'Trade' in df.columns and returns_column == 'Strategy_Returns':
        num_transactions = df['Trade'].sum()
    else:
        num_transactions = np.nan  # Not applicable for benchmark
    
    # Average Gain/Loss per Transaction
    if 'Trade' in df.columns and returns_column == 'Strategy_Returns':
        trades = df[df['Trade'] == 1]
        trade_indices = trades.index
        trade_profits = []
        
        for i in range(1, len(trade_indices)):
            start_index = trade_indices[i-1]
            end_index = trade_indices[i]
            profit = df.loc[end_index, 'Position'] - df.loc[start_index, 'Position']
            trade_profits.append(profit)
        if trade_profits:
            avg_gain_loss = np.mean(trade_profits)
        else:
            avg_gain_loss = 0
    else:
        avg_gain_loss = np.nan  # Not applicable for benchmark
    
    # Maximum Drawdown
    rolling_max = df['Cumulative_Returns'].cummax()
    drawdown = (df['Cumulative_Returns'] - rolling_max) / rolling_max
    max_drawdown = drawdown.min()
    
    # Annualized Sharpe Ratio
    risk_free_rate = 0  # Assuming zero risk-free rate
    periods_per_year = 365.25 * 24 * 60  # Total minutes in a year
    mean_return = df[returns_column].mean() * periods_per_year
    std_return = df[returns_column].std() * np.sqrt(periods_per_year)
    if std_return != 0:
        sharpe_ratio = (mean_return - risk_free_rate) / std_return
    else:
        sharpe_ratio = np.nan
    
    # Tracking Error and Information Ratio (if benchmark returns are provided)
    if benchmark_returns_column:
        df['Excess_Returns'] = df[returns_column] - df[benchmark_returns_column]
        tracking_error = df['Excess_Returns'].std() * np.sqrt(periods_per_year)
        mean_excess_return = df['Excess_Returns'].mean() * periods_per_year
        if tracking_error != 0:
            information_ratio = mean_excess_return / tracking_error
        else:
            information_ratio = np.nan
    else:
        tracking_error = np.nan
        information_ratio = np.nan
    
    # Treynor Ratio and Jensen's Alpha (Require Beta estimation)
    if benchmark_returns_column:
        df.dropna(subset=[returns_column, benchmark_returns_column], inplace=True)
        X = df[benchmark_returns_column]
        y = df[returns_column]
        X = sm.add_constant(X)  # Adds a constant term to the predictor
        model = sm.OLS(y, X).fit()
        beta = model.params[benchmark_returns_column]
        
        if beta != 0:
            treynor_ratio = (mean_return - risk_free_rate) / beta
        else:
            treynor_ratio = np.nan
        mean_benchmark_return = df[benchmark_returns_column].mean() * periods_per_year
        jensens_alpha = mean_return - (risk_free_rate + beta * (mean_benchmark_return - risk_free_rate))
    else:
        beta = np.nan
        treynor_ratio = np.nan
        jensens_alpha = np.nan
    
    ## Metrics Summary ##
    metrics = {
        'Total Return': f"{total_return:.2%}",
        'Annualized Return': f"{annualized_return:.2%}",
        'Number of Transactions': f"{int(num_transactions) if not np.isnan(num_transactions) else 'N/A'}",
        'Average Gain/Loss per Transaction': f"${avg_gain_loss:,.2f}" if not np.isnan(avg_gain_loss) else 'N/A',
        'Maximum Drawdown': f"{max_drawdown:.2%}",
        'Annualized Sharpe Ratio': f"{sharpe_ratio:.2f}",
        'Tracking Error': f"{tracking_error:.2%}" if not np.isnan(tracking_error) else 'N/A',
        'Information Ratio': f"{information_ratio:.2f}" if not np.isnan(information_ratio) else 'N/A',
        'Treynor Ratio': f"{treynor_ratio:.2f}" if not np.isnan(treynor_ratio) else 'N/A',
        'Jensen\'s Alpha': f"{jensens_alpha:.2%}" if not np.isnan(jensens_alpha) else 'N/A'
    }
    
    return metrics

def analyze_strategies(df):
    # Ensure 'Position' and 'Pos_Benchmark' are numeric
    df['Position'] = pd.to_numeric(df['Position'], errors='coerce')
    df['Pos_Benchmark'] = pd.to_numeric(df['Pos_Benchmark'], errors='coerce')
    
    # Calculate Strategy Returns
    df['Strategy_Returns'] = df['Position'].pct_change()
    
    # Calculate Benchmark Returns
    df['Benchmark_Returns'] = df['Pos_Benchmark'].pct_change()
    
    # Metrics for Strategy
    strategy_metrics = calculate_metrics(df.copy(), 'Strategy_Returns', 'Benchmark_Returns')
    
    # Metrics for Benchmark
    benchmark_metrics = calculate_metrics(df.copy(), 'Benchmark_Returns')
    
    # Combine metrics into a DataFrame for better formatting
    metrics_df = pd.DataFrame({
        'Metric': strategy_metrics.keys(),
        'Strategy': strategy_metrics.values(),
        'Benchmark': benchmark_metrics.values()
    })
    
    # Set 'Metric' as the index
    metrics_df.set_index('Metric', inplace=True)
    
    ## Visualizations ##
    # Cumulative Performance Chart
    plt.figure(figsize=(14, 7))
    plt.plot(df.index, (1 + df['Strategy_Returns']).cumprod(), label='Strategy')
    plt.plot(df.index, (1 + df['Benchmark_Returns']).cumprod(), label='Benchmark')
    plt.title('Cumulative Returns')
    plt.xlabel('Date')
    plt.ylabel('Cumulative Returns')
    plt.legend()
    plt.show()
    
    # Returns Distribution for Strategy
    plt.figure(figsize=(10, 5))
    df['Strategy_Returns'].hist(bins=50)
    plt.title('Strategy Returns Distribution')
    plt.xlabel('Returns')
    plt.ylabel('Frequency')
    plt.show()
    
    return metrics_df
```
